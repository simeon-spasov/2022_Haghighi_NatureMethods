### Test for signal in CP and GE

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor

def run_cv(X, y, K, model_choice='ridge', n_inner_folds=5):
    """
    Perform Ridge regression, XGBoost, RandomForest regression, or K-Nearest Neighbors regression
    with K-fold cross-validation, training on 80% of the intervention data and testing on 20% at each iteration.

    Parameters
    ----------
    X : numpy.ndarray
        The input features for the regression.
    y : numpy.ndarray
        The target variable (fitness).
    K : int
        Number of folds for the outer loop of the cross-validation process.
    model_choice : string
        Choice of the model for regression.
    n_inner_folds : int, optional
        Number of folds for the inner loop of the cross-validation process (default is 10).

    Returns
    -------
    dict
        A dictionary containing the cross-validation results, including ridge coefficients,
        normalized root mean square error, and predicted fitness values.
    """
    assert model_choice in ['ridge', 'xgboost', 'randomforest', 'kneighbors'], \
        "Invalid model choice. Choose 'ridge', 'xgboost', 'randomforest', or 'kneighbors'."

    cors = np.zeros((3, K))
    normalised_rmse = np.zeros(K)
    y_test_hat_all = np.zeros_like(y) * np.nan

    # Randomly assign fold indices
    fold_idx_outer = np.random.permutation(np.repeat(np.arange(1, K + 1), len(y) // K + 1)[:len(y)])

    # Set up the outer cross-validation loop
    for k in range(K):
        train_idx = fold_idx_outer != k + 1
        test_idx = fold_idx_outer == k + 1

        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]

        # Choose the model based on the choice
        if model_choice == 'ridge':
            model = RidgeCV(cv=n_inner_folds)
            title = 'Ridge Regression'
        elif model_choice == 'kneighbors':
            parameters = {
                'n_neighbors': range(1, 21)  # hyperoptimize for number of neighbors
            }
            knn_model = KNeighborsRegressor()
            model = GridSearchCV(knn_model, parameters, cv=n_inner_folds)
            title = 'K-Nearest Neighbors Regression'

        model.fit(X_train, y_train)

        # Predict the test set
        y_test_hat = model.predict(X_test)

        y_test_hat_all[test_idx] = y_test_hat

        # Compute correlations and normalized RMSEs
        cors[:, k] = [np.corrcoef(y_test, y_test_hat)[0, 1],
                      pd.Series(y_test).corr(pd.Series(y_test_hat), method='spearman'),
                      pd.Series(y_test).corr(pd.Series(y_test_hat), method='kendall')]

        normalised_rmse[k] = np.sqrt(np.mean((y_test - y_test_hat) ** 2)) / np.std(y)

    # Calculate and plot the line of best fit
    # Here we're plotting this based on the entire dataset to reflect the overall trend

    avg_r_squared = np.mean(cors[0])
    std_r_squared = np.std(cors[0])

    return {'cors': cors,
            'normalised_rmse': normalised_rmse,
            'y_test_hat_all': y_test_hat_all}
```

```python
import Levenshtein

def find_most_similar(target, candidates):
    similarities = [Levenshtein.ratio(target, candidate) for candidate in candidates]
    max_similarity_index = similarities.index(max(similarities))
    return candidates[max_similarity_index], max(similarities)
```

```python
# Extract the first column of broad_compounds where elements start with BRD-
# The full names are like BRD-K76022557-001-03-9, but we truncate to BRD-K76022557, because this is the format used in paper.
# This can lose some information about the compound.
broad_compound_ids = broad_compounds['broad_id'].apply(lambda x: '-'.join(x.split('-')[:2])) 

most_similar_compounds = []

for compound in compounds:
    most_similar, similarity = find_most_similar(compound, broad_compound_ids)
    most_similar_compounds.append((compound, most_similar, similarity))

# Convert to DataFrame for better readability
similarity_df = pd.DataFrame(most_similar_compounds, columns=['Original', 'Most Similar', 'Similarity'])

# Display the results
print(similarity_df)
```

```python
num_perfect_matches = sum(similarity_df["Similarity"]==1)
print(f'Number of perfect matches: {num_perfect_matches}')
print(f'Ratio: {sum(similarity_df["Similarity"]==1)/len(similarity_df["Similarity"])}')
```

```python

# Filter compounds with similarity == 1
perfect_matches = similarity_df[similarity_df['Similarity'] == 1]

assert (perfect_matches['Original'] == perfect_matches['Most Similar']).all()

# Create a list to store the results
detailed_matches = []

# Iterate over the perfect matches
for index, row in perfect_matches.iterrows():
    compound = row['Original']
    
    # Find all broad_compounds that match the compound
    matching_broad_compounds = broad_compounds[broad_compounds['broad_id'].apply(lambda x: '-'.join(x.split('-')[:2])) == compound]
    
    # Extract the SMILES strings for these matches
    smiles_list = matching_broad_compounds['smiles'].tolist()
    
    # Check if all SMILES in the list are the same
    if len(set(smiles_list)) == 1:
        detailed_matches.append((compound, smiles_list[0]))
    print(smiles_list[0])

# Convert to DataFrame for better readability
consistent_smiles_df = pd.DataFrame(detailed_matches, columns=['Compounds', 'SMILES'])

# Print the length of the dataframe
print(f"Number of consistent SMILES matches: {len(consistent_smiles_df)}")

# Display the consistent_smiles dataframe
print(consistent_smiles_df)

print(f"Number of inconsistent SMILES: {num_perfect_matches - len(consistent_smiles_df)}")
```

### Z-score

```python
def z_score(X):
    obs_median = np.median(X, axis=0).reshape(1, -1)
    obs_iqr = np.subtract(*np.percentile(X, [75, 25], axis=0)).reshape(1, -1)
    X = (X - obs_median) / (obs_iqr + 1e-6)
    X = np.clip(X, -10, 10)
    return X
```

```python
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.model_selection import KFold, train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import mean_squared_error 
from scipy.stats import linregress

def find_k_neighbors(G_hat_train, G_hat_sample, k, distance='euclidean'):
    """
    Find k-nearest neighbors in G_hat_train for each sample in G_hat_sample using specified distance.
    """
    neigh = NearestNeighbors(n_neighbors=k, metric=distance)
    neigh.fit(G_hat_train)
    distances, indices = neigh.kneighbors(G_hat_sample)
    return indices

def predict_features_and_targets(X_train, Y_train, indices):
    """
    Predict features and targets by averaging the values from k-nearest neighbors.
    """
    X_pred = np.mean(X_train[indices], axis=1)
    Y_pred = np.mean(Y_train[indices], axis=1)
    return X_pred, Y_pred

def plot_results(all_data, title):
    """
    Plot all results in a single scatter plot, each fold colored differently.
    Calculate and show average line of best fit.
    """
    colors = plt.cm.viridis(np.linspace(0, 1, len(all_data)))
    fig, ax = plt.subplots()
    
    # Collect all data points for calculating global best fit
    all_y_true = np.hstack([data[0] for data in all_data])
    all_y_pred = np.hstack([data[1] for data in all_data])
    
    # Plot each fold with different colors
    for idx, (Y_test, Y_pred) in enumerate(all_data):
        ax.scatter(Y_test, Y_pred, color=colors[idx], label=f'Fold {idx+1}')
    
    # Global line of best fit
    slope, intercept, r_value, p_value, std_err = linregress(all_y_true, all_y_pred)
    ax.plot(all_y_true, intercept + slope * all_y_true, 'k-', label=f'Avg Line of Best Fit: r={r_value:.2f}')
    
    # Calculate correlations and MSEs
    correlations = [np.corrcoef(Y_pred, Y_test)[0, 1] for Y_test, Y_pred in all_data]
    mses = [mean_squared_error(Y_test, Y_pred) for Y_test, Y_pred in all_data]
    mean_corr = np.mean(correlations)
    std_corr = np.std(correlations)
    mean_mse = np.mean(mses)
    std_mse = np.std(mses)
    
    # Displaying correlation and MSE stats
    print(f"{title}: Average Correlation: {mean_corr:.3f} Â± {std_corr:.3f}")
    print(f"{title}: Average MSE: {mean_mse:.3f} Â± {std_mse:.3f}")
    
    ax.legend()
    plt.xlabel('True Y')
    plt.ylabel('Predicted Y')
    plt.title(f'{title}: True vs. Predicted Y with Line of Best Fit Across Folds')
    plt.grid(True)
    plt.show()

def tune_and_predict_ridge(X_train, Y_train, X_val, Y_val, X_test, Y_test):
    """
    Tune Ridge regression parameters using the validation set and predict using the optimal parameters on the test set.
    """
    alphas = np.logspace(-6, 6, 13)
    best_alpha = None
    best_score = float('inf')
    for alpha in alphas:
        ridge = Ridge(alpha=alpha)
        ridge.fit(X_train, Y_train)
        Y_val_pred = ridge.predict(X_val)
        score = mean_squared_error(Y_val, Y_val_pred)
        if score < best_score:
            best_score = score
            best_alpha = alpha
    
    # Retrain with the best alpha and predict on the test set
    ridge = Ridge(alpha=best_alpha)
    ridge.fit(X_train, Y_train)
    Y_test_pred = ridge.predict(X_test)
    return Y_test, Y_test_pred

def perform_cross_validation(X, G_hat, Y, k=5, n_splits=5, plot=False, random_state=42):
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=random_state)
    all_data_kNN = []
    all_data_ridge_on_X_pred = []
    all_data_ridge_combined = []
    all_data_ridge_on_G_hat = []

    for train_index, test_index in kf.split(X):
        print('Processing fold...')        
        X_train_full, X_test = X[train_index], X[test_index]
        G_hat_train_full, G_hat_test = G_hat[train_index], G_hat[test_index]
        Y_train_full, Y_test = Y[train_index], Y[test_index]

        X_train, X_val, Y_train, Y_val, G_hat_train, G_hat_val = train_test_split(
            X_train_full, Y_train_full, G_hat_train_full, test_size=0.2, random_state=random_state)

        # KNN to predict X_val_pred for validation tuning
        indices_val = find_k_neighbors(G_hat_train, G_hat_val, k=k, distance=distance)
        X_val_pred, _ = predict_features_and_targets(X_train, Y_train, indices_val)

        # KNN to predict X_test_pred for final prediction
        indices_test = find_k_neighbors(G_hat_train_full, G_hat_test, k=k, distance=distance)
        X_test_pred, Y_test_pred_kNN = predict_features_and_targets(X_train_full, Y_train_full, indices_test)
        all_data_kNN.append((Y_test, Y_test_pred_kNN))

        print(f"MSE between X_test_pred and X_test is: {mean_squared_error(X_test, X_test_pred)}")

        # Ridge on X_pred features only
        Y_test, Y_test_pred_ridge_X_pred = tune_and_predict_ridge(
            X_train, Y_train, X_val_pred, Y_val, X_test_pred, Y_test)
        all_data_ridge_on_X_pred.append((Y_test, Y_test_pred_ridge_X_pred))

        # Ridge on G_hat features only
        Y_test, Y_test_pred_ridge_G_hat = tune_and_predict_ridge(
        G_hat_train, Y_train, G_hat_val, Y_val, G_hat_test, Y_test)
        all_data_ridge_on_G_hat.append((Y_test, Y_test_pred_ridge_G_hat))

        # Ridge on combined G_hat and X_pred features
        combined_features_train = np.concatenate((G_hat_train, X_train), axis=1)
        combined_features_val = np.concatenate((G_hat_val, X_val_pred), axis=1)
        combined_features_test = np.concatenate((G_hat_test, X_test_pred), axis=1)
        Y_test, Y_test_pred_combined = tune_and_predict_ridge(
            combined_features_train, Y_train, combined_features_val, Y_val, combined_features_test, Y_test)
        all_data_ridge_combined.append((Y_test, Y_test_pred_combined))
    
    if plot:
        plot_results(all_data_kNN, 'kNN on G_hat')
        plot_results(all_data_ridge_on_X_pred, 'Ridge on predicted X')
        plot_results(all_data_ridge_on_G_hat, 'Ridge on G_hat')
        plot_results(all_data_ridge_combined, 'Ridge on G_hat and predicted X')

# The rest of your implementation remains unchanged
perform_cross_validation(X, G_hat, Y, k=5, plot=True)
```

```python
import numpy as np
import pickle
import os
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
import torch.nn.functional as F
import copy

class Encoder(nn.Module):
    def __init__(self, feature_dim):
        super().__init__()
        self.mlp = model = nn.Sequential(
            nn.Linear(feature_dim, feature_dim),
            nn.ReLU(),
            nn.Linear(feature_dim, feature_dim),
            nn.ReLU(),
        )
    
    def forward(self, x):
        return self.mlp(x)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class AdaptiveNPairLoss(nn.Module):
    def __init__(self, top_k_percent=5, num_neg_samples=10):
        super(AdaptiveNPairLoss, self).__init__()
        self.top_k_percent = top_k_percent
        self.num_neg_samples = num_neg_samples

    def forward(self, embeddings, targets):
        n = embeddings.size(0)
        k = max(1, int(self.top_k_percent / 100 * n))  # number of top k closest neighbors

        # Compute pairwise distances in target space
        target_diffs = torch.abs(targets.unsqueeze(0) - targets.unsqueeze(1))

        # Get the indices of the top-k closest neighbors for each sample in target space
        _, top_k_indices = torch.topk(target_diffs, k=k+1, dim=1, largest=False)  # +1 because the first one is itself
        
        # Ensure the first element in each row is the index itself
        for i in range(n):
            if top_k_indices[i, 0] != i:
                # Find the position of i in top_k_indices[i]
                pos = (top_k_indices[i] == i).nonzero(as_tuple=True)[0].item()
                # Swap the first element with the element at position pos
                top_k_indices[i, pos] = top_k_indices[i, 0]
                top_k_indices[i, 0] = i

        # Remove self (first column) from top_k_indices
        top_k_indices = top_k_indices[:, 1:k+1]

        # Randomly sample a positive for each row
        rand_idx = torch.randint(0, k, (n,))
        positive_indices = top_k_indices[torch.arange(n), rand_idx]

        # Calculate Euclidean distance between anchor and positive
        pos_distances = torch.norm(embeddings - embeddings[positive_indices], dim=1)

        # Calculate Euclidean distances between anchor and all samples
        all_distances = torch.cdist(embeddings, embeddings, p=2)

        # Mask to exclude positive samples and self-distances for negatives
        mask = torch.ones_like(all_distances, dtype=torch.bool)
        mask[torch.arange(n), positive_indices] = False
        mask[torch.arange(n), torch.arange(n)] = False

        # Ensure the mask has the correct number of False entries
        masked_sum = mask.sum(dim=-1)
        assert (masked_sum == (n - 2)).all(), f"Expected {(n - 2)}, got {masked_sum}"

        # Calculate negative distances
        neg_distances = all_distances[mask].view(n, n - 2)

        # Randomly sample num_neg_samples negatives for each anchor
        if self.num_neg_samples < (n - 2):
            sampled_neg_indices = torch.multinomial(torch.ones(n - 2), self.num_neg_samples, replacement=False)
            neg_distances = neg_distances[:, sampled_neg_indices]

        # Compute softmax with positive distance as numerator and sum of sampled negative distances as denominator
        pos_distances_exp = torch.exp(-pos_distances).unsqueeze(1)
        neg_distances_exp = torch.exp(-neg_distances)
        
        # Normalize positive distances with sampled negative distances
        numerator = pos_distances_exp
        denominator = (pos_distances_exp + neg_distances_exp.sum(dim=1, keepdim=True))

        # print(numerator)
        # print(denominator)
        
        softmax_loss = -torch.log(numerator / denominator)

        return softmax_loss.mean()

# Example usage
top_k_percent = 5
num_neg_samples = 10  # Define the number of negative samples to use
loss_fn = AdaptiveNPairLoss(top_k_percent, num_neg_samples)

embeddings = torch.rand(10, 128)  # Example embeddings
targets = torch.rand(10)          # Example continuous targets

loss = loss_fn(embeddings, targets)
print(loss)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class ScalableTripletLoss(nn.Module):
    def __init__(self, top_k_percent=5, margin=1.0):
        super(ScalableTripletLoss, self).__init__()
        self.top_k_percent = top_k_percent
        self.margin = margin

    def forward(self, embeddings, targets):
        n = embeddings.size(0)
        k = max(1, int(self.top_k_percent / 100 * n))  # number of top k closest neighbors

        # Compute pairwise distances in target space
        target_diffs = torch.abs(targets.unsqueeze(0) - targets.unsqueeze(1))

        # Compute pairwise Euclidean distances in embedding space
        embedding_diffs = torch.cdist(embeddings, embeddings, p=2)

        # Get the indices of the top-k closest neighbors for each sample in target space
        _, top_k_indices = torch.topk(target_diffs, k=k+1, dim=1, largest=False)  # +1 because the first one is itself
        
        # Create a mask to exclude self-comparisons
        mask = torch.ones_like(target_diffs)
        mask.fill_diagonal_(0)

        # Sample one positive pair from top-k neighbors for each sample
        top_k_indices = top_k_indices[:, 1:]  # Remove self
        pos_indices = top_k_indices[:, torch.randint(k, (1,)).repeat(n, 1)].squeeze()

        # Randomly sample one negative pair for each sample
        all_indices = torch.arange(n).unsqueeze(0).repeat(n, 1)
        neg_indices = torch.masked_select(all_indices, mask.bool()).view(n, n-1)
        neg_indices = neg_indices[:, torch.randint(n-1, (1,)).repeat(n, 1)].squeeze()

        # Calculate triplet loss
        anchor_embeddings = embeddings
        positive_embeddings = embeddings[pos_indices]
        negative_embeddings = embeddings[neg_indices]

        pos_distances = F.pairwise_distance(anchor_embeddings, positive_embeddings)
        neg_distances = F.pairwise_distance(anchor_embeddings, negative_embeddings)

        triplet_loss = F.relu(pos_distances - neg_distances + self.margin)
        
        return triplet_loss.mean()

# Example usage
top_k_percent = 5
margin = 1.0
loss_fn = ScalableTripletLoss(top_k_percent, margin)

embeddings = torch.rand(10, 128)  # Example embeddings
targets = torch.rand(10)          # Example continuous targets

loss = loss_fn(embeddings, targets)
print(loss)
```
